```python
import pandas as pd
train_df = pd.read_json("../inputs/training.txt",orient='records', lines=True)
```

```python
train_df.shape
```

```python
train_df.drop_duplicates(subset=['内容'], inplace=True)
```

```python
vali_df = pd.read_json("../inputs/validation.txt",orient='records', lines=True)
```

```python
vali_df
```

```python
import jieba
clean_content = lambda text: " ".join(" ".join(jieba.cut(text)).split())
```

```python
clean_content("据媒体报道，中国西南四川省的古佛像被斩首后，警方正在抓头。")
```

```python
train_df['内容'] = train_df['内容'].apply(clean_content)
```

```python
vali_df['内容'] = vali_df['内容'].apply(clean_content)
```

```python
train_df.to_csv("../inputs/train.tsv", index=False, sep='\t')
vali_df.to_csv("../inputs/vali.tsv", index=False, sep='\t')
```
